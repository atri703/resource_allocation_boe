```python
def get_new_column_names(df):
    '''
    Input - DataFrame
    Output - DataFrame
    '''
    # Get the values from the first row as a list
    first_row_values = df.iloc[0].tolist()

    # Extract the existing column names
    existing_column_names = df.columns.tolist()
    existing_column_names = [col.rstrip('.1234567890') for col in existing_column_names]

    # Concatenate the first row values with the existing column names
    new_column_names = [f"{col} - {val}" for col, val in zip(existing_column_names, first_row_values)]

    # Rename the columns of the DataFrame with the new column names
    df.columns = new_column_names
    
    df = df.rename(columns={'Firm name - nan': 'Firm name'})
    
    # Dropping first row
    df = df.drop(0)
    
    return df
```

```python
def get_float_type(df):
    
    df_metrics = df.drop(['Firm name'], axis=1)
    
    df_metrics = df_metrics.applymap(lambda x: float(x) if pd.notnull(x) else x)
    
    df = pd.concat([df['Firm name'], df_metrics], axis=1)
    
    return df
```

```python
df_general = get_new_column_names(df_general)
df_underwriting = get_new_column_names(df_underwriting)
```
