```python
def get_new_column_names(df):
    '''
    Input - DataFrame
    Output - DataFrame
    '''
    # Get the values from the first row as a list
    first_row_values = df.iloc[0].tolist()

    # Extract the existing column names
    existing_column_names = df.columns.tolist()
    existing_column_names = [col.rstrip('.1234567890') for col in existing_column_names]

    # Concatenate the first row values with the existing column names
    new_column_names = [f"{col} - {val}" for col, val in zip(existing_column_names, first_row_values)]

    # Rename the columns of the DataFrame with the new column names
    df.columns = new_column_names
    
    df = df.rename(columns={'Firm name - nan': 'Firm name'})
    
    # Dropping first row
    df = df.drop(0)
    
    return df
```

```python
def get_float_type(df):
    
    df_metrics = df.drop(['Firm name'], axis=1)
    
    df_metrics = df_metrics.applymap(lambda x: float(x) if pd.notnull(x) else x)
    
    df = pd.concat([df['Firm name'], df_metrics], axis=1)
    
    return df
```

## 1. Finding firms based on Firm Sizes

### Appproach:

#### 1. Identify 30 biggest firms based on GWP and NWP
#### 2. Sort them on the basis on mean values
#### 3. Out of 30 firms again filter to find those firms which shows substantial change year to year.

```python
# Table of Firm Sizes for NWP
firm_sizes = final_df_general[['Firm name', 'NWP (£m)  - 2020YE', 'NWP (£m)  - 2019YE', 'NWP (£m)  - 2018YE', 'NWP (£m)  - 2017YE', 'NWP (£m)  - 2016YE']]

firm_sizes['Mean NWP'] = firm_sizes.mean(axis=1)

firm_sizes = firm_sizes.sort_values(by='Mean NWP', ascending=False)

print("Table of Firm Sizes:")
firm_sizes
```

```python
# Table of Firm Sizes for NWP
firm_sizes_gwp = final_df_general[['Firm name', 'GWP (£m) - 2020YE', 'GWP (£m) - 2019YE', 'GWP (£m) - 2018YE', 'GWP (£m) - 2017YE', 'GWP (£m) - 2016YE']]

firm_sizes_gwp['Mean GWP'] = firm_sizes_gwp.mean(axis=1)

firm_sizes_gwp = firm_sizes_gwp.sort_values(by='Mean GWP', ascending=False)

print("Table of Firm Sizes:")
firm_sizes_gwp
```

```python
# Step 1: Assuming 'firm_sizes' and 'firm_sizes_gwp' already contain the average calculations and are sorted
top_30_nwp_firms = firm_sizes.head(30)
top_30_gwp_firms = firm_sizes_gwp.head(30)

# Step 2: Calculate Year-to-Year Growth for NWP and GWP within the top 30 firms
for year in range(2016, 2020):
    year_growth_nwp = f'NWP Growth {year}-{year+1}'
    top_30_nwp_firms[year_growth_nwp] = (top_30_nwp_firms[f'NWP (£m)  - {year+1}YE'] - top_30_nwp_firms[f'NWP (£m)  - {year}YE']) / top_30_nwp_firms[f'NWP (£m)  - {year}YE']

for year in range(2016, 2020):
    year_growth_gwp = f'GWP Growth {year}-{year+1}'
    top_30_gwp_firms[year_growth_gwp] = (top_30_gwp_firms[f'GWP (£m) - {year+1}YE'] - top_30_gwp_firms[f'GWP (£m) - {year}YE']) / top_30_gwp_firms[f'GWP (£m) - {year}YE']

# Handle infinite values due to division by zero
top_30_nwp_firms.replace([np.inf, -np.inf], np.nan, inplace=True)
top_30_gwp_firms.replace([np.inf, -np.inf], np.nan, inplace=True)

# Step 3: Identify Firms with Significant Year-to-Year Changes in NWP and GWP
# Define "significant" as changes beyond a certain threshold, for simplicity, use the interquartile range (25th-75th percentile) as the threshold
threshold_nwp = top_30_nwp_firms[[f'NWP Growth {year}-{year+1}' for year in range(2016, 2020)]].quantile([0.25, 0.75])
threshold_gwp = top_30_gwp_firms[[f'GWP Growth {year}-{year+1}' for year in range(2016, 2020)]].quantile([0.25, 0.75])

significant_changes_nwp = top_30_nwp_firms[
    ((top_30_nwp_firms[[f'NWP Growth {year}-{year+1}' for year in range(2016, 2020)]] < threshold_nwp.loc[0.25]) |
    (top_30_nwp_firms[[f'NWP Growth {year}-{year+1}' for year in range(2016, 2020)]] > threshold_nwp.loc[0.75])).any(axis=1)
]

significant_changes_gwp = top_30_gwp_firms[
    ((top_30_gwp_firms[[f'GWP Growth {year}-{year+1}' for year in range(2016, 2020)]] < threshold_gwp.loc[0.25]) |
    (top_30_gwp_firms[[f'GWP Growth {year}-{year+1}' for year in range(2016, 2020)]] > threshold_gwp.loc[0.75])).any(axis=1)
]

# Extract firm names with significant changes
firms_nwp_significant_changes = significant_changes_nwp.index.tolist()
firms_gwp_significant_changes = significant_changes_gwp.index.tolist()

# Combine and deduplicate firm names
all_firms_with_significant_changes = list(set(firms_nwp_significant_changes + firms_gwp_significant_changes))

all_firms_with_significant_changes
```

```python
# Gross Claims Incurred: Identify firms with substantial fluctuations in gross claims incurred
gross_claims_cols = final_df_underwriting.filter(regex='Gross claims incurred').columns

final_df_underwriting[gross_claims_cols] = final_df_underwriting[gross_claims_cols].apply(pd.to_numeric, errors='coerce')

gross_claims_year_to_year_changes = final_df_underwriting[gross_claims_cols].pct_change(axis=1).abs()

firms_with_substantial_claims_changes = final_df_underwriting[(gross_claims_year_to_year_changes.max(axis=1) > 10.0) & final_df_underwriting['Gross claims incurred (£m) - 2020YE'] > 0]
```
